```python
from string import punctuation
from nltk.corpus import stopwords
from nltk import word_tokenize
from nltk.stem import SnowballStemmer
```

```python
df['genre'] = df['genre'].astype(str)
```

```python
def clean_txt(sent):    

    
    tokens = word_tokenize(sent.lower())
    stop_updated =list(punctuation) + ["would", "could","told"]
    x = [term for term in tokens if term not in stop_updated ] 
    res = " ".join(x)
    return res
```

```python
df['ctext'] = df.genre.apply(clean_txt)
```

```python
df['cname'] = df.name.apply(clean_txt)
```

```python
df
```

```python
X_text = df.ctext.values
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf_vectorizer = TfidfVectorizer(max_features=1000)
X = tfidf_vectorizer.fit_transform(X_text)
```

```python
X.toarray()
```

```python
from sklearn.mixture import GaussianMixture
gmm = GaussianMixture(n_components=500)
gmm.fit(X.toarray())
labels = gmm.predict(X.toarray())
```
